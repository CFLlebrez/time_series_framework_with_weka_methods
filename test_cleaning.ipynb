## Data cleaning notebook
Este notebook se utilizará para limpieza de ejemplo de un csv.

```python
import pandas as pd

osuna = pd.read_csv("input_csv_files/Osuna.csv")

print(osuna.head())
```

```python
print(f"Campos vacíos:\n{osuna.isnull().sum()}") #Ver campos vacíos (no hay)
print(f"Filas duplicadas: {osuna.duplicated().sum()}") # Ver el número de duplicados (no hay)
```

```python
print(f"Número de valores distintos:\n{osuna.nunique()}")
```

```python
columnas = osuna.columns
print(osuna.head())
```

```python
# Hay que convertir las columnas a numéricas
for col in columnas:
    if col not in ['FECHA', 'Se11HTMax', 'Se11HTMin']:
        osuna[col] = pd.to_numeric(osuna[col], errors='coerce')
# FECHA a datetime y Se11HTMax y Se11HTMin a time
osuna['FECHA'] = pd.to_datetime(osuna['FECHA'], errors='coerce', format='%d/%m/%y')
```

```python
osuna = osuna.drop(labels=['Se11HTMax', 'Se11HTMin'], axis=1) #columnas innecesarias
```

```python
osuna.loc[:,~osuna.columns.str.contains('^Unnamed')] #columna extra
```

```python
osuna.dropna(inplace=True) #valores vacíos, dado que en este caso hay muchos datos se dropean directamente.
```

```python
osuna.to_csv('osuna_clean.csv')
```
